```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([2,4,8,16,32,64])
iteration_num_rastrigin = np.array([20, 243, 854, 2877, 2992, 4200])
iteration_num_rozenbrock = np.array([52, 467, 1122, 1822, 3111, 3888])
iteration_num_ra_mod = np.array([30, 213, 722, 1890, 2720, 3328])
iteration_num_ro_mod = np.array([70, 117, 520, 1120, 2899, 3129])
```

```python
plt.figure(figsize=(10,6))
plt.plot(X, iteration_num_rastrigin, c='b', label="ABC")
plt.plot(X, iteration_num_ra_mod, c='r', label="Модифицированный ABC")
plt.scatter(X, iteration_num_rastrigin, c='b', s=20)
plt.scatter(X, iteration_num_ra_mod, c='r', s=20)
plt.grid()
plt.legend()
plt.figure
plt.xlabel("|X|")
plt.ylabel(r"$\bart$", loc="top", rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(X, iteration_num_rozenbrock, c='b', label="ABC")
plt.plot(X, iteration_num_ro_mod, c='r', label="Модифицированный ABC")
plt.scatter(X, iteration_num_rozenbrock, c='b', s=20)
plt.scatter(X, iteration_num_ro_mod, c='r', s=20)
plt.grid()
plt.legend()
plt.figure
plt.xlabel("|X|")
plt.ylabel(r"$\bart$", loc="top", rotation=0)
plt.show()
```
